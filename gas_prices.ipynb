```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(10,6))

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Germany']

for country in list_country:
    plt.plot(gas['Year'], gas[country], marker='.', label=country)

plt.title('Valor do combustível em US$')
plt.xlabel('Anos')
plt.ylabel('Dólares')

plt.xticks(gas['Year'][::2], rotation=45)

plt.savefig('contry.png')
plt.legend()
plt.show()

```
